```python
from mediabyte import omm
```

## Playback

#### Browser and VLC Playback

Yota, Cue and Sample objects can be viewed in a browser tab with `play()`:

```python
omm('y.R2ipPgrWypI.10m.11m').play()
```

or viewed in VLC with vlc():

```python
omm('y.R2ipPgrWypI.10m.11m').vlc()
```

Remembering that `omm()` can parse YouTube URLs, we can use these directly:

```python
omm('https://www.youtube.com/watch?v=dZLyfbSQPXI').vlc()
```

A Mixtape folds out into individual browser tabs with `Mixtape.open_tabs()`:

```python
omm('y.R2ipPgrWypI.10m.11m..y.R2ipPgrWypI.13m.14m').open_tabs()
```

#### Mixtape HTML Player

A HTML/JavaScript Mixtape player can be written to file using `Mixtape.write_player_html(filename)`:

```python
omm('y.R2ipPgrWypI.10m.11m..y.R2ipPgrWypI.13m.14m').write_player_html('mixtapeplayer.html')
```

Or the HTML accessed directly with `Mixtape.player()`:

```python
omm('y.R2ipPgrWypI.10m.11m..y.R2ipPgrWypI.13m.14m').player()
```
